```python
query = r"""
        WITH base AS (
            SELECT
                TPRECOSVEN_IT.ID AS PRECO_FOCCO_ID,
                TITENS.COD_ITEM,
                TPRECOSVEN.COD_PREVEN,
                TPRECOSVEN.DESCRICAO AS TABELA_DESCRICAO,
                REGEXP_REPLACE(TITENS.DESC_TECNICA, '^MODELO\s+', '', 'i') AS PRODUTO,
        
                TCARACTERISTICAS.COD_CAR,
                TVARIAVEIS.MNEMONICO,
                TITENS_CAR.SEQ,
        
                TPRECOSVEN_IT.DES_FORMULA AS FORMULA,
                TPRECOSVEN_IT.PRECO AS PRECO
        
            FROM TPRECOSVEN_IT
        
            JOIN TPRECOSVEN
                ON TPRECOSVEN.ID = TPRECOSVEN_IT.TPRVEN_ID
        
            JOIN TITENS_COMERCIAL
                ON TITENS_COMERCIAL.ID = TPRECOSVEN_IT.ITCM_ID
        
            JOIN TITENS_EMPR
                ON TITENS_EMPR.ID = TITENS_COMERCIAL.ITEMPR_ID
        
            JOIN TITENS
                ON TITENS.ID = TITENS_EMPR.ITEM_ID
        
            LEFT JOIN TPRC_REGRAS_COM
                ON TPRC_REGRAS_COM.TPRVEN_IT_ID = TPRECOSVEN_IT.ID
        
            LEFT JOIN TCARACTERISTICAS
                ON TCARACTERISTICAS.ID = TPRC_REGRAS_COM.TCARAC_ID
        
            LEFT JOIN TITENS_CAR
                ON TITENS_CAR.ITEMPR_ID = TITENS_EMPR.ID
               AND TITENS_CAR.TCARAC_ID = TPRC_REGRAS_COM.TCARAC_ID
        
            LEFT JOIN TPRC_REGRAS_VAR_COM
                ON TPRC_REGRAS_VAR_COM.REGRA_COM_ID = TPRC_REGRAS_COM.ID
        
            LEFT JOIN TVARIAVEIS
                ON TVARIAVEIS.ID = TPRC_REGRAS_VAR_COM.TVAR_ID
        
            WHERE TPRECOSVEN_IT.SIT = 1
              --AND TITENS.COD_ITEM = '20001'
              AND TPRECOSVEN.COD_PREVEN = 155
        )
        
        SELECT
            PRECO_FOCCO_ID,
            COD_ITEM,
            COD_PREVEN,
            TABELA_DESCRICAO,
            PRODUTO,
        
            MAX(MNEMONICO) FILTER (WHERE COD_CAR = 'MODULACAO') AS MODULACAO,
            MAX(MNEMONICO) FILTER (WHERE COD_CAR = 'COMP_MODULO') AS COMP_MODULO,
            MAX(MNEMONICO) FILTER (WHERE COD_CAR = 'PROF_PRODUTO') AS PROF_PRODUTO,
        
            REPLACE(
                MAX(MNEMONICO) FILTER (WHERE COD_CAR = 'FX_TEC'),
                'FX ',
                ''
            ) AS FAIXA,
        
            MAX(MNEMONICO) FILTER (WHERE COD_CAR = 'TIPO_ACAB') AS TIPO_ACAB,
            MAX(MNEMONICO) FILTER (WHERE COD_CAR = 'EMBAL_REFORCADA') AS EMBALAGEM,
        
            STRING_AGG(
                MNEMONICO,
                CHR(35)
                ORDER BY SEQ
            ) FILTER (WHERE MNEMONICO IS NOT NULL) AS CONFIGURACAO,
        
            FORMULA,
        
            STRING_AGG(
                COD_CAR || ':' || MNEMONICO,
                CHR(124)
                ORDER BY SEQ
            ) FILTER (
                WHERE COD_CAR IS NOT NULL
                  AND MNEMONICO IS NOT NULL
            ) AS RESP,
        
            PRECO
        
        FROM base
        
        GROUP BY
            PRECO_FOCCO_ID,
            COD_ITEM,
            COD_PREVEN,
            TABELA_DESCRICAO,
            PRODUTO,
            FORMULA,
            PRECO
        
        ORDER BY
            COD_ITEM,
            PRODUTO,
            MODULACAO,
            COMP_MODULO,
            PROF_PRODUTO,
            FAIXA,
            TIPO_ACAB,
            EMBALAGEM;

        """

```

```python
df = pd.read_sql(query, engine)

engine.dispose()
```

```python
import xmlrpc.client
import pandas as pd

# =====================================
# ODOO
# =====================================
url = "https://crm.brainess.com.br"
db = "sohome_18"
username = "tiago@sohome.com"
password = "admin"

common = xmlrpc.client.ServerProxy(f"{url}/xmlrpc/2/common")
uid = common.authenticate(db, username, password, {})
if not uid:
    raise Exception("Falha na autenticacao")

models = xmlrpc.client.ServerProxy(f"{url}/xmlrpc/2/object")
MODEL = "calculadora.price.line"
print(f"Conectado como uid={uid}")

# =====================================
# CONFIG -- altere antes de importar
# =====================================
MARCA = "CENTURY"   # nome exato da marca no Odoo

RESP_KEYS_COVERED = {
    "MODULACAO", "COMP_MODULO", "PROF_PRODUTO",
    "FX_TEC", "TIPO_ACAB", "EMBAL_REFORCADA",
}
RESP_KEYS_EXTRAS_CONHECIDAS = {
    "BASE_PE", "TIPO_MESA", "PECA_XADREZ", "FORMATO_MOD", "ALT_MODULO", "MATERIAL_TAMPO",
    "TP_ALMO", "TP_ENCHIMENTO", "FX_CAIXA", "ACESSORIO", "MAT_PE_BASE", "TP_ACAB_AP",
    "COR_VIVO", "FX_TEC_VIVO", "PROF_PROD", "COMP_BRACO", "TIPO_CAMA", "TP_ACAB_ACESS",
    "TIPO_KIT_MOT", "QTD_AS", "TIPO_ACAB_BDJ", "TP_ACAB_TAMPO", "COMP_CONSOLE",
    "FORMATO_CONSOLE", "OPCAO_BDJ", "OPCAO_TAPECADO", "OPCAO_MANCEBO", "COMP_ASS",
    "TIPO_ACAB_DET", "TIPO_CONFORTO", "VAR_ENCOSTO",
}

# =====================================
# DATAFRAME
# =====================================
df_import = df.copy()
df_import.columns = [str(c).strip().upper() for c in df_import.columns]

required_cols = [
    "PRECO_FOCCO_ID", "COD_ITEM", "COD_PREVEN", "TABELA_DESCRICAO", "PRODUTO",
    "MODULACAO", "COMP_MODULO", "PROF_PRODUTO", "FAIXA", "TIPO_ACAB", "EMBALAGEM",
    "CONFIGURACAO", "FORMULA", "RESP", "PRECO",
]
missing = [c for c in required_cols if c not in df_import.columns]
if missing:
    raise Exception(f"Colunas ausentes no dataframe: {missing}")

total = len(df_import)
print(f"Total de linhas no dataframe: {total}")

# =====================================
# FUNCOES DE LIMPEZA
# =====================================
def clean_str(v):
    return "" if pd.isna(v) else str(v).strip()

def clean_int(v):
    if pd.isna(v) or str(v).strip() == "":
        return None
    try:
        return int(float(v))
    except (ValueError, TypeError):
        return None

def clean_float(v):
    return 0.0 if pd.isna(v) or str(v).strip() == "" else float(v)

def get_or_create(model, domain, values):
    ids = models.execute_kw(db, uid, password, model, "search", [domain])
    if ids:
        return ids[0]
    return models.execute_kw(db, uid, password, model, "create", [values])

# =====================================================================
# DATA HEALTH CHECKS
# =====================================================================
print("\n" + "="*60)
print("DATA HEALTH CHECKS")
print("="*60)

mask_preco_ruim = df_import["PRECO"].isna() | (df_import["PRECO"] <= 0)
n_preco_ruim = mask_preco_ruim.sum()
print(f"\n[1] Preco <= 0 ou nulo          : {n_preco_ruim:>6}")
if n_preco_ruim:
    print(df_import[mask_preco_ruim][["PRODUTO", "COD_ITEM", "FAIXA", "PRECO"]].drop_duplicates().head(10).to_string())

dupes_mask = df_import["PRECO_FOCCO_ID"].duplicated(keep=False)
n_dupes = dupes_mask.sum()
print(f"\n[2] preco_focco_id duplicado    : {n_dupes:>6}")
if n_dupes:
    print(df_import[dupes_mask][["PRECO_FOCCO_ID", "PRODUTO", "FAIXA", "PRECO"]].head(10).to_string())

mask_sem_produto = df_import["PRODUTO"].isna() | (df_import["PRODUTO"].str.strip() == "")
print(f"\n[3] Produto vazio               : {mask_sem_produto.sum():>6}")

mask_unico = df_import["FAIXA"].isna() | (df_import["FAIXA"].str.strip() == "")
n_unico = mask_unico.sum()
pct_unico = n_unico / total * 100 if total else 0
print(f"\n[4] Faixa vazia (-> UNICO)      : {n_unico:>6}  ({pct_unico:.1f}%)")
if n_unico:
    print("    Produtos com mais linhas UNICO:")
    print(df_import[mask_unico]["PRODUTO"].str.strip().str.upper().value_counts().head(15).to_string())

mask_sem_mod  = df_import["MODULACAO"].isna()   | (df_import["MODULACAO"].str.strip()   == "")
mask_sem_comp = df_import["COMP_MODULO"].isna() | (df_import["COMP_MODULO"].str.strip() == "")
mask_sem_prof = df_import["PROF_PRODUTO"].isna()| (df_import["PROF_PRODUTO"].str.strip()== "")
mask_so_extras = mask_sem_mod & mask_sem_comp & mask_sem_prof
n_so_extras = mask_so_extras.sum()
print(f"\n[5] Sem modulacao/medida (so extras): {n_so_extras:>6}")
if n_so_extras:
    print(df_import[mask_so_extras]["PRODUTO"].str.strip().str.upper().value_counts().head(20).to_string())

todas_chaves = set()
for resp_str in df_import["RESP"].dropna():
    for part in str(resp_str).split("|"):
        if ":" in part:
            todas_chaves.add(part.split(":")[0].strip())
chaves_extras_encontradas = todas_chaves - RESP_KEYS_COVERED
chaves_novas = chaves_extras_encontradas - RESP_KEYS_EXTRAS_CONHECIDAS
print(f"\n[6] Chaves extras no resp       : {len(chaves_extras_encontradas):>6}")
print(f"    Todas: {sorted(chaves_extras_encontradas)}")
if chaves_novas:
    print(f"\n    *** ATENCAO: Chaves NOVAS nao reconhecidas: {sorted(chaves_novas)} ***")
    print("    Adicione ao EXTRAS_LABELS em app.js e ao RESP_KEYS_EXTRAS_CONHECIDAS acima.")
else:
    print("    Nenhuma chave nova detectada.")

def tem_extra_no_resp(resp_str):
    if not resp_str or pd.isna(resp_str):
        return False
    chaves = {part.split(":")[0].strip() for part in str(resp_str).split("|") if ":" in part}
    return bool(chaves - RESP_KEYS_COVERED)

mask_sem_extra_resp = ~df_import["RESP"].apply(tem_extra_no_resp)
mask_sem_dif = mask_unico & mask_so_extras & mask_sem_extra_resp
n_sem_dif = mask_sem_dif.sum()
print(f"\n[7] Sem qualquer diferenciacao  : {n_sem_dif:>6}")
if n_sem_dif:
    print(df_import[mask_sem_dif][["PRODUTO", "PRECO_FOCCO_ID", "RESP", "PRECO"]].head(10).to_string())

print(f"\n[8] Top 15 produtos por linhas  :")
print(df_import.groupby("PRODUTO").size().sort_values(ascending=False).head(15).to_string())

print("\n" + "="*60)
print("FIM DOS HEALTH CHECKS")
print("="*60 + "\n")

# =====================================
# PASSO 1 -- CATEGORIAS
# =====================================
categoria_ids = {}
if "CATEGORIA" in df_import.columns:
    print("Sincronizando categorias...")
    for cat in df_import["CATEGORIA"].dropna().unique():
        cat = str(cat).strip()
        if not cat:
            continue
        cid = get_or_create("product.category", [["name", "=", cat]], {"name": cat})
        categoria_ids[cat] = cid
        print(f"   '{cat}' -> id={cid}")

# =====================================
# PASSO 2 -- MARCA
# =====================================
print("\nSincronizando marca...")
brand_id = get_or_create("product.brand", [["name", "=", MARCA]], {"name": MARCA})
print(f"   '{MARCA}' -> id={brand_id}")

# =====================================
# PASSO 3 -- PRODUTOS (product.template)
# Cria se nao existe, atualiza se ja existe.
# =====================================
print("\nSincronizando produtos...")

produtos_df = (
    df_import[["PRODUTO", "COD_ITEM"] + (["CATEGORIA"] if "CATEGORIA" in df_import.columns else [])]
    .drop_duplicates(subset=["PRODUTO"])
)

criados = existentes = erros_prod = 0

for _, row in produtos_df.iterrows():
    nome   = clean_str(row["PRODUTO"])
    codigo = clean_str(row["COD_ITEM"])
    cat    = clean_str(row["CATEGORIA"]) if "CATEGORIA" in row.index else ""

    if not nome:
        continue

    domain = ["|", ["default_code", "=", codigo], ["name", "=", nome]] if codigo \
             else [["name", "=", nome]]
    ids_exist = models.execute_kw(db, uid, password, "product.template", "search", [domain])

    vals = {
        "name":         nome,
        "default_code": codigo,
        "type":         "consu",
        "sale_ok":      True,
        "purchase_ok":  False,
        "brand_id":     brand_id,
    }
    if cat and cat in categoria_ids:
        vals["categ_id"] = categoria_ids[cat]

    try:
        if ids_exist:
            models.execute_kw(db, uid, password, "product.template", "write", [ids_exist, vals])
            existentes += 1
        else:
            new_id = models.execute_kw(db, uid, password, "product.template", "create", [vals])
            criados += 1
            print(f"   Criado: {nome} (id={new_id})")
    except Exception as e:
        erros_prod += 1
        print(f"   Erro em '{nome}': {e}")

print(f"Produtos -> criados={criados}  atualizados={existentes}  erros={erros_prod}")

# =====================================
# PASSO 4 -- LINHAS DE PRECO
# =====================================
print("\nImportando linhas de preco...")

existing_records = models.execute_kw(
    db, uid, password, MODEL, "search_read",
    [[["preco_focco_id", "!=", False]]],
    {"fields": ["preco_focco_id"]}
)
existing_ids = {r["preco_focco_id"] for r in existing_records}
print(f"   Ja cadastrados: {len(existing_ids)}")

def row_to_vals(row):
    faixa_raw   = clean_str(row["FAIXA"])
    faixa       = faixa_raw.replace("FX ", "").strip() or "UNICO"
    marca_linha = clean_str(row["MARCA"]) if "MARCA" in row.index else ""
    cat_linha   = clean_str(row["CATEGORIA"]) if "CATEGORIA" in row.index else ""
    return {
        "active":           True,
        "preco_focco_id":   clean_int(row["PRECO_FOCCO_ID"]),
        "cod_item":         clean_str(row["COD_ITEM"]),
        "cod_preven":       clean_int(row["COD_PREVEN"]) or 0,
        "tabela_descricao": clean_str(row["TABELA_DESCRICAO"]),
        "produto":          clean_str(row["PRODUTO"]),
        "marca":            marca_linha or MARCA,
        "categoria":        cat_linha,
        "modulacao":        clean_str(row["MODULACAO"]),
        "comp_modulo":      clean_str(row["COMP_MODULO"]),
        "prof_produto":     clean_str(row["PROF_PRODUTO"]),
        "faixa":            faixa,
        "faixa_raw":        faixa_raw,
        "tipo_acab":        clean_str(row["TIPO_ACAB"]),
        "embalagem":        clean_str(row["EMBALAGEM"]),
        "configuracao":     clean_str(row["CONFIGURACAO"]),
        "formula":          clean_str(row["FORMULA"]),
        "resp":             clean_str(row["RESP"]),
        "preco":            clean_float(row["PRECO"]),
    }

rows_to_insert   = []
skipped_sem_id   = []
skipped_sem_prod = []
skipped_preco    = []
skipped_dupes    = 0

for _, row in df_import.iterrows():
    vals      = row_to_vals(row)
    produto   = vals["produto"]
    focco_id  = vals["preco_focco_id"]
    vals.pop("faixa_raw")

    if focco_id is None:
        skipped_sem_id.append({"produto": produto})
        continue
    if not produto:
        skipped_sem_prod.append({"preco_focco_id": focco_id})
        continue
    if vals["preco"] <= 0:
        skipped_preco.append({"produto": produto, "preco_focco_id": focco_id, "preco": vals["preco"]})
        continue
    if focco_id in existing_ids:
        skipped_dupes += 1
        continue
    rows_to_insert.append(vals)

print(f"\n{'='*50}")
print(f"Sem preco_focco_id : {len(skipped_sem_id):>6}")
print(f"Sem produto        : {len(skipped_sem_prod):>6}")
print(f"Preco <= 0         : {len(skipped_preco):>6}")
print(f"Ja no Odoo (normal): {skipped_dupes:>6}")
print(f"A inserir          : {len(rows_to_insert):>6}")
print(f"{'='*50}")

if skipped_sem_id:
    print("\n Linhas sem preco_focco_id:")
    print(pd.DataFrame(skipped_sem_id).head(20).to_string())
if skipped_preco:
    print("\n Linhas com preco <= 0:")
    print(pd.DataFrame(skipped_preco).head(20).to_string())

BATCH_SIZE = 100
created = 0
errors  = []

try:
    for i in range(0, len(rows_to_insert), BATCH_SIZE):
        batch = rows_to_insert[i : i + BATCH_SIZE]
        try:
            models.execute_kw(db, uid, password, MODEL, "create", [batch])
            created += len(batch)
            print(f"  {created}/{len(rows_to_insert)} inseridos...")
        except Exception:
            for vals in batch:
                try:
                    models.execute_kw(db, uid, password, MODEL, "create", [vals])
                    created += 1
                except Exception as e2:
                    errors.append({"preco_focco_id": vals.get("preco_focco_id"),
                                   "produto": vals.get("produto"), "erro": str(e2)})
except KeyboardInterrupt:
    print("\nInterrompido")
finally:
    print(f"\n{'='*50}")
    print(f"Criados : {created}")
    print(f"Erros   : {len(errors)}")
    if errors:
        print(pd.DataFrame(errors).head(20).to_string())
```
