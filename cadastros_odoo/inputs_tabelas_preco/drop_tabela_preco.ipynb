```python
#script dropar tabela
import xmlrpc.client

url      = "https://crm.brainess.com.br"
db       = "sohome_18"
username = "tiago@sohome.com"
password = "admin"

common = xmlrpc.client.ServerProxy(f"{url}/xmlrpc/2/common")
uid    = common.authenticate(db, username, password, {})
models = xmlrpc.client.ServerProxy(f"{url}/xmlrpc/2/object")
print(f"✅ Conectado como uid={uid}")

# =====================================
# ALTERE AQUI
# =====================================
COD_PREVEN = 25  # código da tabela a dropar

# =====================================
# APAGA LINHAS DE PREÇO
# =====================================
ids = models.execute_kw(db, uid, password,
    "calculadora.price.line", "search",
    [[["cod_preven", "=", COD_PREVEN]]]
)
print(f"Linhas encontradas: {len(ids)}")

if ids:
    models.execute_kw(db, uid, password, "calculadora.price.line", "unlink", [ids])
    print(f"✅ {len(ids)} linhas removidas (cod_preven={COD_PREVEN})")
else:
    print("Nenhuma linha encontrada.")

# =====================================
# APAGA O REGISTRO DA TABELA
# =====================================
table_ids = models.execute_kw(db, uid, password,
    "calculadora.price.table", "search",
    [[["cod_preven", "=", COD_PREVEN]]]
)
if table_ids:
    models.execute_kw(db, uid, password, "calculadora.price.table", "unlink", [table_ids])
    print(f"✅ Tabela cod_preven={COD_PREVEN} removida de calculadora.price.table")
```
