```python
import requests

BASE_URL = "https://ollama.brainess.com.br"
TOKEN = "tokentiagonovelli"

headers = {
    "Authorization": f"Bearer {TOKEN}",
    "Content-Type": "application/json",
}

payload = {
    "prompt": "quais as suas principais habilidades? voce é capaz de criar agentes ou skills?"
}
```

```python

response = requests.post(
    f"{BASE_URL}/chat",
    headers=headers,
    json=payload,
    timeout=120,
)

print(response.status_code)
print(response.json())
```
