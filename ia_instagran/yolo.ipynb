```python
from ultralytics import YOLO 
import cv2 # baixa/carrega um modelo pré-treinado 
model = YOLO("yolov8n.pt") 
# imagem 
#img_path = r"C:\Users\tiago.premiano\Downloads\sala.jpg" 
img_path = r"C:\Users\tiago.premiano\Downloads\sala3.jpeg" 
# detectar objetos 
results = model(img_path) 
# desenhar caixas na imagem 
annotated = results[0].plot() 
# mostrar resultado 
cv2.imshow("Detecção", annotated) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```
