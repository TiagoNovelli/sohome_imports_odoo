```python
from ultralytics import YOLO
import cv2

# baixa/carrega um modelo pré-treinado
model = YOLO("yolov8n.pt")

# imagem
img_path = r"C:\Users\tiago.premiano\Downloads\sala.jpg"

#img_path = "https://i0.wp.com/blog.divinahaus.com.br/wp-content/uploads/2017/07/23.158_Blog-min.png?w=1200&ssl=1"
img = cv2.imread(img_path)

img = cv2.imwrite(r"C:\Users\tiago.premiano\Downloads\sala4.jpeg", img)

# detectar objetos
results = model(img_path)

# desenhar caixas na imagem
annotated = results[0].plot()

# mostrar resultado
cv2.imshow("Detecção", annotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
