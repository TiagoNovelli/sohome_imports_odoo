```python
def df_to_js(dataframe: object, output_path: str):
    df = dataframe
    group_keys = ["produto_x", "modulacao", "descricao", "comprimento", "profundidade", "acabamento", "embalagem"]
 
    # Preenche NaN nas chaves com string vazia para não perder linhas no groupby
    df_g = df.copy()
    for col in group_keys:
        df_g[col] = df_g[col].fillna("").astype(str).str.strip()
 
    produtos = {}
 
    for keys, group in df_g.groupby(group_keys, sort=False):
        produto_x, modulacao, descricao, comprimento, profundidade, acabamento, embalagem = keys
 
        comp = comprimento if comprimento else ""
        prof = profundidade if profundidade else ""
        if comp and prof:
            med = f"C:{comp} × P:{prof}"
        elif comp:
            med = f"C:{comp}"
        else:
            med = ""
 
        precos = {}
        for _, row in group.iterrows():
            faixa = str(row["faixa"]).strip() if pd.notna(row["faixa"]) else None
            faixa_caixa = str(row["faixa_caixa"]).strip() if pd.notna(row["faixa_caixa"]) else None
            preco = row["preco"]
 
            chave = faixa if faixa and faixa != "nan" else faixa_caixa
            if chave and chave != "nan" and pd.notna(preco):
                precos[chave] = float(preco)
 
        entry = {
            "m": modulacao,
            "med": med,
            "acab": acabamento,
            "emb": embalagem,
            "d": descricao,
            "$": precos,
        }
 
        if produto_x not in produtos:
            produtos[produto_x] = []
        produtos[produto_x].append(entry)
 
    # ── 2. MARCAS (produto → marca comercial, ex: CENTURY, PONTOVIRGULA) ─────
    marcas_dict = (
        df[["produto_x", "marca"]]
        .drop_duplicates(subset=["produto_x"])
        .sort_values("produto_x")
        .pipe(lambda d: {
            str(r["produto_x"]).strip(): str(r["marca"]).strip()
            for _, r in d.iterrows()
            if pd.notna(r["produto_x"]) and pd.notna(r["marca"])
        })
    )
 
    # ── 3. CATEGORIAS (produto → classificação, ex: SOFAS LIVING, POLTRONAS) ─
    categorias_dict = (
        df[["produto_x", "classificacao"]]
        .drop_duplicates(subset=["produto_x"])
        .sort_values("produto_x")
        .pipe(lambda d: {
            str(r["produto_x"]).strip(): str(r["classificacao"]).strip()
            for _, r in d.iterrows()
            if pd.notna(r["produto_x"]) and pd.notna(r["classificacao"])
        })
    )
 
    # ── 4. CLASSIFICACOES (lista ordenada única de categorias) ────────────────
    classificacoes = sorted(df["classificacao"].dropna().unique().tolist())
 
    # ── 5. PRECISA CAIXA ─────────────────────────────────────────────────────
    tem_caixa = set(df[df["embalagem"] == "CAIXA MADEIRA"]["produto_x"].dropna().unique())
    tem_sem_caixa = set(df[df["embalagem"] == "SEM CX MADEIRA"]["produto_x"].dropna().unique())
    precisa_caixa = sorted([str(p).strip() for p in tem_caixa & tem_sem_caixa])
 
    # ── 6. TEM FAIXA CAIXA ───────────────────────────────────────────────────
    tem_faixa_caixa = sorted(
        str(p).strip()
        for p in df[df["faixa_caixa"].notna() & (df["faixa_caixa"].astype(str) != "")]["produto_x"]
        .dropna()
        .unique()
    )
 
    # ── MONTA O JS ───────────────────────────────────────────────────────────
    data = {
        "produtos": produtos,
        "marcas": marcas_dict,
        "categorias": categorias_dict,
        "classificacoes": classificacoes,
        "precisaCaixa": precisa_caixa,
        "temFaixaCaixa": tem_faixa_caixa,
    }
 
    json_str = json.dumps(data, ensure_ascii=False, separators=(",", ":"))
    js_output = f"const D={json_str};\n;DATA=D;render();\n"
 
    with open(output_path, "w", encoding="utf-8") as f:
        f.write(js_output)
 
    import os
    size_mb = os.path.getsize(output_path) / 1024 / 1024
    print(f"✅ Gerado: {output_path} ({size_mb:.2f} MB)")
    print(f"   Produtos: {len(produtos)}, Marcas: {len(marcas_dict)}, Categorias: {len(categorias_dict)}")
    print(f"   Classificações: {len(classificacoes)}")
    print(f"   precisaCaixa: {len(precisa_caixa)}, temFaixaCaixa: {len(tem_faixa_caixa)}")
 
```
